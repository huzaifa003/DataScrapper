```python
import nltk
nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
import pandas as pd
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
```

```python
mobile_page = requests.get("https://www.whatmobile.com.pk/Samsung_Mobiles_Prices")
soup = BeautifulSoup(mobile_page.content, 'lxml')
mobile_div = soup.find("div", {"class" : "mobiles"})
mobile_div
```

```python
mobiles = mobile_div.find_all("div", {"class" : "item"})
mobiles
```

```python
data = {"title" : [], "price_PKR" : []}
print(data['title'])
for mobile in mobiles:
    title = mobile.find("a", {"class" : "BiggerText"}).text.strip()
    
    title = title.replace("Samsung", "Samsung ")
    data['title'].insert(0,title);

    price = mobile.find("span", {"class" : "PriceFont"}).text.strip();
    data['price_PKR'].insert(0,price)

data    
```

```python
df = pd.DataFrame(data);
df
```

```python
df['price_PKR'].replace(["Price N/A", "N/A"],pd.NA, inplace=True) # Making N/A Values
df
```

```python
df.dropna(how="any", inplace=True) #Dropping NAN Values
```

```python
df.drop_duplicates()
df
```

```python
def wordlist(text):
    blob = TextBlob(text)
    return blob.tags
```

```python
df['tags'] = df['title'].apply(wordlist)
df
```

```python
def priceCleaner(text):
    text = text.replace("Rs. ", "")
    text = text.replace(",","")
    return text
```

```python
df['price_PKR'] = df['price_PKR'].apply(priceCleaner);
```
